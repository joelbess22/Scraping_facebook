```python
data = data.loc[:,~data.columns.str.match("Unnamed")] 
data.drop(['imageMaisonLien'],axis=1,inplace=True) 
data.rename(columns={'localite':'commune'}, inplace=True)
data.drop(['username'],axis=1,inplace=True)
```

```python

def montant(data):
    for i in range(data.shape[0]):
        if  len(str(data['montant'].iloc[i])) ==1:
            data['montant'][i] = data['montant'][i]*100000
        if len(str(data['montant'].iloc[i])) == 2:
            data['montant'][i] = data['montant'][i]*10000
        if  len(str(data['montant'].iloc[i])) ==3:
            data['montant'][i] = data['montant'][i]*1000
        if  len(str(data['montant'].iloc[i])) ==4:
            data['montant'][i] = data['montant'][i]*100
        if  len(str(data['montant'].iloc[i])) ==5:
            data['montant'][i] = data['montant'][i]*10
    return data
data=montant(data)
```

```python
mediane = data['nbPiece'].median()
data['nbPiece'].fillna(mediane, inplace=True)
```

```python
# Analyse par localité de type

# grouper par "localite" et "type"
localite_type = data.groupby(['commune', 'type'])

# calculer le montant minimal, le montant moyen et le montant maximal pour chaque groupe
stats = localite_type['montant'].agg(['min', 'mean', 'max'])

# afficher les résultats
print(stats)
```

```python
#Analyse par localité et par type
def par_localite_type(data):
    # grouper par "localite" et "type"
    localite_type = data.groupby(['commune', 'type'])

    # calculer le montant minimal, le montant moyen et le montant maximal pour chaque groupe
    stats = localite_type['montant'].agg(['min', 'mean', 'max'])

    return stats
par_localite_type(data)
```

```python
# Nombre de_Pièces et calcul du montant min,max,mean 
def groupe_nbPiece(data):
    groupe = data.groupby(['commune', 'nbPiece'])
    stats = groupe['montant'].agg(['min', 'mean', 'max'])
    stats.plot(kind='bar',color=['red', 'green', 'blue'])
    plt.show()
groupe_nbPiece(data)

 
```

```python
#Repartition des maisons en fonction du nombre de pieces en fonction de l'annéé
def groupby_nbPiece_annee(data):
    grouped = data.groupby(['annee'])
    stats = grouped['montant'].agg(['min', 'mean', 'max'])
    stats.plot(kind='bar',color=['red', 'green', 'blue'])
    plt.show()
groupby_nbPiece_annee(data)
 
```
